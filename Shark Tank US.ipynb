```python
shark_tank_df.columns = shark_tank_df.columns.str.strip().str.replace(' ', '_').str.lower()
```

```python
shark_tank_df.columns
```

```python
shark_tank_df.isnull().sum()
```

```python
missing_cols = ['pitchers_city', 'pitchers_state', 'pitchers_average_age', 'entrepreneur_names', 'company_website']
shark_tank_df[missing_cols].isnull().mean() * 100
```

```python
if 'pitchers_city' in shark_tank_df.columns:
    shark_tank_df.drop(columns=['pitchers_city'], inplace=True)
```

```python
shark_tank_df['pitchers_state'] = shark_tank_df['pitchers_state'].fillna('Unknown')
```

```python
shark_tank_df['pitchers_average_age'].head()
```

```python
# Check unique values in the 'pitchers_average_age' column to identify any non-numeric data
shark_tank_df['pitchers_average_age'].unique()
```

```python
age_map = {
    'Young': 20,  # Example: assign Young as 20
    'Middle': 40, # Example: assign Middle as 40
    'Old': 60     # Example: assign Old as 60
}
```

```python
shark_tank_df['pitchers_average_age'] = shark_tank_df['pitchers_average_age'].map(age_map)
```

```python
mean_age = shark_tank_df['pitchers_average_age'].mean()
shark_tank_df['pitchers_average_age'] = shark_tank_df['pitchers_average_age'].fillna(mean_age)
```

```python
shark_tank_df['pitchers_average_age'].head()
```

```python
shark_tank_df['pitchers_average_age'].dtype
```

```python
shark_tank_df['entrepreneur_names'] = shark_tank_df['entrepreneur_names'].fillna('Unknown')
```

```python
shark_tank_df['company_website'] = shark_tank_df['company_website'].fillna('Not Available')
```

```python
shark_tank_df['deal_made'] = shark_tank_df['got_deal'].astype(bool)
```

```python
deal_columns = [
    'total_deal_amount', 'total_deal_equity', 'deal_valuation',
    'number_of_sharks_in_deal', 'investment_amount_per_shark', 'equity_per_shark',
    'royalty_deal', 'loan'
]
```

```python
print(shark_tank_df[['royalty_deal', 'loan']].dtypes)
```

```python
# Fill NaN values in 'royalty_deal' and 'loan' directly with 'No Deal'
shark_tank_df['royalty_deal'] = shark_tank_df['royalty_deal'].fillna('No Deal')
shark_tank_df['loan'] = shark_tank_df['loan'].fillna('No Deal')
```

```python
# Check if there are any NaN values left
print(shark_tank_df[['royalty_deal', 'loan']].isnull().sum())
```

```python
# Fill numerical columns with 0
shark_tank_df.loc[~shark_tank_df['deal_made'], deal_columns] = \
    shark_tank_df.loc[~shark_tank_df['deal_made'], deal_columns].fillna(0)
```

```python
# Fill string columns separately
# Replace 0 with 'No Deal' for clarity  I used direct assignment
shark_tank_df['royalty_deal'] = shark_tank_df['royalty_deal'].replace(0, 'No Deal')
shark_tank_df['loan'] = shark_tank_df['loan'].replace(0, 'No Deal')
```

```python
shark_columns = [col for col in shark_tank_df.columns if '_investment_' in col]
```

```python
# Fill all shark investment columns with 0 where no deal
shark_tank_df.loc[~shark_tank_df['deal_made'], shark_columns] = 0
```

```python
shark_tank_df['guest_name'] = shark_tank_df['guest_name'].fillna('No Guest')
shark_tank_df['guest_investment_amount'] = shark_tank_df['guest_investment_amount'].fillna(0)
shark_tank_df['guest_investment_equity'] = shark_tank_df['guest_investment_equity'].fillna(0)
```

```python
presence_cols = [col for col in shark_tank_df.columns if '_present' in col]

for col in presence_cols:
    shark_tank_df[col] = shark_tank_df[col].fillna(False).astype(bool)
```

```python
shark_tank_df['pitchers_gender'] = shark_tank_df['pitchers_gender'].fillna('NA')
```

```python
shark_tank_df['multiple_entrepreneurs'] = shark_tank_df['multiple_entrepreneurs'].fillna(0.0)
```

```python
# Fill the missing values with 0 for investment amount and equity columns
shark_tank_df['barbara_corcoran_investment_amount'] = shark_tank_df['barbara_corcoran_investment_amount'].fillna(0)
shark_tank_df['barbara_corcoran_investment_equity'] = shark_tank_df['barbara_corcoran_investment_equity'].fillna(0)

shark_tank_df['mark_cuban_investment_amount'] = shark_tank_df['mark_cuban_investment_amount'].fillna(0)
shark_tank_df['mark_cuban_investment_equity'] = shark_tank_df['mark_cuban_investment_equity'].fillna(0)

shark_tank_df['lori_greiner_investment_amount'] = shark_tank_df['lori_greiner_investment_amount'].fillna(0)
shark_tank_df['lori_greiner_investment_equity'] = shark_tank_df['lori_greiner_investment_equity'].fillna(0)

shark_tank_df['robert_herjavec_investment_amount'] = shark_tank_df['robert_herjavec_investment_amount'].fillna(0)
shark_tank_df['robert_herjavec_investment_equity'] = shark_tank_df['robert_herjavec_investment_equity'].fillna(0)

shark_tank_df['daymond_john_investment_amount'] = shark_tank_df['daymond_john_investment_amount'].fillna(0)
shark_tank_df['daymond_john_investment_equity'] = shark_tank_df['daymond_john_investment_equity'].fillna(0)

shark_tank_df['kevin_o_leary_investment_amount'] = shark_tank_df['kevin_o_leary_investment_amount'].fillna(0)
shark_tank_df['kevin_o_leary_investment_equity'] = shark_tank_df['kevin_o_leary_investment_equity'].fillna(0)
```
